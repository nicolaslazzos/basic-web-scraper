```python
import pandas as pd
```

### Reading the file

```python
df = pd.read_csv('el_universal-2020_10_31-articles.csv')
df
```

### Making some transfotmation

Adding the site_id to the dataframe

```python
df['site_id'] = 'el_universal'

df
```

Get the host of each news

```python
from urllib.parse import urlparse
```

```python
df['host'] = df['url'].apply(lambda url: urlparse(url).netloc)
df
```

```python
df['host'].value_counts()
```

### Dealing with missing data

```python
missing_titles_mask = df['title'].isna()

missing_titles = df[missing_titles_mask]['url'].str.extract(r'(?P<missing_titles>[^/]+)$').applymap(lambda title: title.replace('-', ' '))

df.loc[missing_titles_mask, 'title'] = missing_titles.loc[:, 'missing_titles']

df
```

### Setting custom ids

```python
import hashlib
```

```python
uids = df.apply(lambda row: hashlib.md5(bytes(row['url'].encode())).hexdigest(), axis=1)
df['uid'] = uids
df.set_index('uid', inplace=True)

df
```

### Cleaning stuff

```python
stripped_body = df.apply(lambda row: row['body'].replace('\n', '').replace('\r', ''), axis=1)
df['body'] = stripped_body
stripped_body
```

```python
import nltk
from nltk.corpus import stopwords
```

```python
stop_words = set(stopwords.words('spanish'))

def tokenize_column(df, column_name):
    return (df.apply(lambda row: nltk.word_tokenize(row[column_name]), axis=1)
            .apply(lambda tokens: list(filter(lambda token: token.isalpha(), tokens)))
            .apply(lambda tokens: list(map(lambda token: token.lower(), tokens)))
            .apply(lambda words_list: len(list(filter(lambda word: word not in stop_words, words_list)))))

df['n_tokens_title'] = tokenize_column(df, 'title')
df['n_tokens_body'] = tokenize_column(df, 'body')
df
```

### Duplicated values

```python
df['title'].value_counts()
df.drop_duplicates(subset=['title'], keep='first', inplace=True)
df.drop_duplicates(subset=['url'], keep='first', inplace=True)
```
