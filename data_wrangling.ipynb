```python
import pandas as pd
```

### Reading the file

```python
df = pd.read_csv('el_universal_2020_10_31_articles.csv')
df
```

### Making some transfotmation

Adding the site_id to the dataframe

```python
df['site_id'] = 'el_universal'

df
```

Get the host of each news

```python
from urllib.parse import urlparse
```

```python
df['host'] = df['url'].apply(lambda url: urlparse(url).netloc)
df
```

```python
df['host'].value_counts()
```

### Dealing with missing data

```python
missing_titles_mask = df['title'].isna()

missing_titles = df[missing_titles_mask]['url'].str.extract(r'(?P<missing_titles>[^/]+)$').applymap(lambda title: title.replace('-', ' '))

df.loc[missing_titles_mask, 'title'] = missing_titles.loc[:, 'missing_titles']

df
```

### Setting custom ids

```python
import hashlib
```

```python
uids = df.apply(lambda row: hashlib.md5(bytes(row['url'].encode())).hexdigest(), axis=1)
df['uid'] = uids
df.set_index('uid', inplace=True)

df
```

### Cleaning stuff

```python
stripped_body = df.apply(lambda row: row['body'].replace('\n', '').replace('\r', ''), axis=1)
stripped_body
```
